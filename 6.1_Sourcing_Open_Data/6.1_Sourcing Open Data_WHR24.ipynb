### World Happiness Report 2024

#### Load data

```python
import pandas as pd
import numpy as np
import os

# Define paths
base_path = r'C:\Users\moein\anaconda3\06_World Happiness Report 2024\02 Data'
input_path = os.path.join(base_path, 'Original Data')
output_path = os.path.join(base_path, 'Prepared Data')

# Define file path
file_path_1 = os.path.join(input_path, "World-happiness-report-2024.csv")
file_path_2 = os.path.join(input_path, "World-happiness-report-updated_2024.csv")

# Load the CSV into a DataFrame
df_WHR = pd.read_csv(file_path_1, encoding='latin1')
df_WHR_updated = pd.read_csv(file_path_2, encoding='latin1')

# Display basic information
df_WHR.info()
df_WHR.head()
```

### Check Overall Structure

```python
# Check structure and get a feel for the columns
df_WHR_updated.info()
df_WHR_updated.head()
df_WHR_updated.describe(include='all')  # Includes categorical summaries
```

### Drop Unnecessary Columns

```python
df_WHR_updated.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
```

### Check for Missing Values

```python
# Total missing per column
print(df_WHR_updated.isnull().sum())
```

#### Impute missing values

```python
# Fill numeric missing values with column means
columns_to_fill = [
    'Log GDP per capita',
    'Social support',
    'Healthy life expectancy at birth',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption',
    'Positive affect',
    'Negative affect'
]

for col in columns_to_fill:
    df_WHR_updated[col] = df_WHR_updated[col].fillna(df_WHR_updated[col].mean())
```

### Check for Duplicates

```python
print("Duplicates:", df_WHR_updated.duplicated().sum())
df_WHR_updated.drop_duplicates(inplace=True)
```

### Check and Fix Data Types

```python
# Convert year column to integer
df_WHR_updated['year'] = pd.to_numeric(df_WHR_updated['year'], errors='coerce')

# Check again
df_WHR_updated.dtypes
```

### Standardize Categorical Values

```python
df_WHR_updated['Country name'] = df_WHR_updated['Country name'].str.strip()
```

### Get a Basic Profile

```python
print("Shape:", df_WHR_updated.shape)
print("Years covered:", df_WHR_updated['year'].unique())
print("Number of countries:", df_WHR_updated['Country name'].nunique())
```

### Save the Cleaned Dataset

```python
# Save the cleaned dataset
cleaned_file_path = os.path.join(output_path, "Cleaned_World_Happiness_2024.csv")
df_WHR_updated.to_csv(cleaned_file_path, index=False)
```
